```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# First create some toy data:  
x = np.linspace(0, 2*np.pi, 400)  
y = np.sin(x**2)  
```

```python

# Creates just a figure and only one subplot  
fig, ax = plt.subplots()  
ax.plot(x, y)  
ax.set_title('Simple plot')  
```

```python
# Creates two subplots and unpacks the output array immediately  
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 1))  
# fig是大图，ax1和ax2代表小图,一行两列
# 2列1行，分享y轴，大小10*10英寸
ax1.plot(x,y)  
ax1.set_title('Sharing Y axis') 
ax1.set_xlabel("X")
#图1标题 , X轴名称
ax2.scatter(x, y)  
ax2.set_title('Sharing Y axis') 
ax2.set_ylabel("Y", rotation=0)
#图2是散点图，Y轴名称，旋转轴刻度上文字方向
```

```python
# Creates four polar axes, and accesses them through the returned array  
fig, axes = plt.subplots(2, 2, subplot_kw=dict(polar=True), figsize = (10,10))  
# fig是大图，axes是小图，用axes[x,y]的形式画小图
axes[0, 0].plot(x, y)  
axes[1, 1].scatter(x, y)  
```

```python
# method 1: subplot2grid
##########################
plt.figure(figsize=(10,10))
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)  
# 用subplot2grid画子图1
ax1.plot([1, 2], [1, 2])
ax1.set_title('ax1_title')
#子图设置标题要加个set_title
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=1)
ax2.plot([1, 2], [1, 2])
ax3 = plt.subplot2grid((3, 3), (1, 1), colspan=1)
ax3.plot([1, 2], [1, 2])
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4.plot([1, 2], [1, 2])
ax5 = plt.subplot2grid((3, 3), (2, 0))
ax5.scatter([1, 2], [2, 2])
ax5.set_xlabel('ax4_x')
ax5.set_ylabel('ax4_y')
ax6 = plt.subplot2grid((3, 3), (2, 1))
ax6.plot([1, 2], [1, 2])
```

```python
# method 2: gridspec
#########################
import matplotlib.gridspec as gridspec
plt.figure(figsize=(10,10))
gs = gridspec.GridSpec(3, 3)
# 这里相当于subplot2grid的第一个参数
# use index from 0 从0开始计算grid的序数
ax1 = plt.subplot(gs[0, :])
ax1.plot([1, 2], [1, 2])
ax2 = plt.subplot(gs[1, :2])
ax2.plot([1, 2], [1, 2])
ax3 = plt.subplot(gs[1:, 2])
ax3.plot([1, 2], [1, 2])
ax4 = plt.subplot(gs[-1, 0])
ax4.plot([1, 2], [1, 2])
ax5 = plt.subplot(gs[-1, -2])
ax5.plot([1, 2], [1, 2])
```

```python
# method 3: easy to define structure
####################################
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7,ax8,ax9)) = plt.subplots(3, 3, sharex=True, sharey=True, figsize=(10,10))
ax1.plot([1, 2], [1, 2])
ax1.set_label('cool pic 1')
ax2.plot([1, 2], [1, 2])
ax3.plot([1, 2], [1, 2])
ax4.plot([1, 2], [1, 2])
ax5.plot([1, 2], [1, 2])
ax6.plot([1, 2], [1, 2])
ax7.plot([1, 2], [1, 2])
ax8.plot([1, 2], [1, 2])
ax9.plot([1, 2], [1, 2])
plt.tight_layout()
plt.subplots_adjust(top=0.85)
# tight_layout会自动调整子图参数，使之填充整个图像区域。
plt.show()
```

```python
plt.figure(figsize=(6, 4))
# plt.subplot(n_rows, n_cols, plot_num)
# 先造出大图，再用subplot（三个数字）来画小图
plt.subplot(221)
plt.plot([0, 1], [0, 1])

plt.subplot(222)
plt.plot([0, 1], [0, 2])

plt.subplot(223)
plt.plot([0, 1], [0, 3])

plt.subplot(224)
plt.plot([0, 1], [0, 4])

plt.tight_layout()
```

```python
plt.figure(figsize=(6, 4))
# plt.subplot(n_rows, n_cols, plot_num)
plt.subplot(2, 1, 1)
# figure splits into 2 rows, 1 col, plot to the 1st sub-fig
plt.plot([0, 1], [0, 1])

plt.subplot(234)
# figure splits into 2 rows, 3 col, plot to the 4th sub-fig
plt.plot([0, 1], [0, 2])

plt.subplot(235)
# figure splits into 2 rows, 3 col, plot to the 5th sub-fig
plt.plot([0, 1], [0, 3])

plt.subplot(236)
# figure splits into 2 rows, 3 col, plot to the 6th sub-fig
plt.plot([0, 1], [0, 4])


plt.tight_layout()
plt.show()
```
