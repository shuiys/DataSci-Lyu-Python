```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 50)
y = 2*x + 1
# y = x**2
```

```python
plt.plot(x, y)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x, y)
ax.set_title('Simple plot')
ax.set_xlabel("X")  
ax.set_ylabel("Y", rotation=0)
```

```python
x = np.linspace(-3, 3, 50)
# -3到3的50个数字组成的等差数列
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y1)

plt.figure(num=3, figsize=(4, 4))
# num=3 代表图片的序号定义为figure 3
# figsize代表图片的长和高（英寸）
plt.plot(x, y2)
# plot the second curve in this figure with certain parameters
plt.plot(x, y1, color='red', linewidth=1.5, linestyle='--')
# color 是线的颜色
# linewidth是线的宽度
# linestyle是线的形态
plt.show()
```

```python
'-'       solid line style
'--'      dashed line style
'-.'      dash-dot line style
':'       dotted line style
```
