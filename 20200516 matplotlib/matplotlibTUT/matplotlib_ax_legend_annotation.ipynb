```python
# -*- coding: UTF-8 -*-
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

创建数据

```python
plt.figure()
plt.plot(x, y2)
# plot the second curve in this figure with certain parameters
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
# set x limits
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.title('cool pic')
plt.xlabel('I am x')
plt.ylabel('I am y')
```

画y1和y2两个函数
设定x和y轴的界限
设置x和y轴的label

```python
# set new sticks
plt.figure(figsize=(5, 5))
plt.plot(x, y2)
# plot the second curve in this figure with certain parameters
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
# set x limits
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
# set tick labels
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
print('xticks下标现在要定义为：', new_ticks)
print('yticks下标现在要定义为：', ['really bad', 'bad', 'normal', 'good', 'really good'])
plt.show()
```

xticks来设置x轴的小标
yticks来设置y轴的小标

```python
# set new sticks
plt.figure(figsize=(5, 5))
plt.plot(x, y2)
# plot the second curve in this figure with certain parameters
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
# set x limits
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
# set tick labels
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
# to use '$ $' for math text and nice looking, e.g. '$\pi$'

# gca = 'get current axis'
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
print('现在上面和右边的轴颜色就变了，用set_color(\'none\')则可以隐藏轴')

ax.xaxis.set_ticks_position('bottom')
# ACCEPTS: [ 'top' | 'bottom' | 'both' | 'default' | 'none' ]
ax.spines['bottom'].set_position(('data', 0))
# the 1st is in 'outward' | 'axes' | 'data'
# axes: percentage of y axis
# data: depend on y data 定位在y轴数据的0点那里

ax.yaxis.set_ticks_position('left')
# ACCEPTS: [ 'left' | 'right' | 'both' | 'default' | 'none' ]
ax.spines['left'].set_position(('data',0))
# 定位在x轴数据的0点那里
print('调整ticks在x轴和y轴的位置，并且坐标轴相交在零点')

plt.show()
```

用gca = 'get current axis'获取当前坐标轴
并用set_ticks_position和spines.set_position来调整坐标轴的位置

```python
# set new sticks
plt.figure(figsize=(7, 7))
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
# set tick labels
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
# to use '$ $' for math text and nice looking, e.g. '$\pi$'

l1, = plt.plot(x, y1, color='blue', label='label 1', linestyle='-')
l2, = plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--', label='label 2')
plt.legend(handles = [l1,l2,], labels= ['up', 'down'], loc='lower right') # loc就是legend location图例位置的意思
# plt.legend(handles=[l1, l2], labels=['up', 'down'],  loc='best')
# the "," is very important in here l1, = plt... and l2, = plt... for this step
plt.show()
```

legend和一些小技巧

```python
y = 2*x + 1
plt.figure(num=1, figsize=(8, 8),)
plt.plot(x, y,)
plt.yticks([-5,-4,-3,-2,-1, 1, 2, 3, 4, 5])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#消除top和right的轴颜色
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
#让x轴和y轴在零点相交

x0 = 1
y0 = 2*x0 + 1
plt.scatter(x0, y0, s=50, color='b')
plt.plot([x0, x0,], [y0, 0,], 'k--', color = 'gold', linewidth=2.5)
#在图上画一个散点，并在坐标轴上作垂直线


# method 1:
#####################
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xycoords='data', xytext=(+30, -40),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

# method 2:
########################
plt.text(-3, 3, r'$ some\ symbols.\ \mu\ \sigma_i\ \alpha_t $',
         fontdict={'size': 16, 'color': 'deeppink'})
#反斜杠可以输入α
plt.show()
```

在函数旁边加图注

```python
y = 0.1*x

plt.figure(figsize=(8, 8))
plt.plot(x, y, linewidth=3, zorder=1)      
# set zorder for ordering the plot in plt 2.0.2 or higher
# zorder是为了判断不同函数在z轴的相对位置
plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
#调整坐标轴以零点为中心

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16) #设置tick label的字体大小
    # set zorder for ordering the plot in plt 2.0.2 or higher
    label.set_bbox(dict(facecolor='green', edgecolor='gold', alpha=0.2, zorder=2)) 
    #设置tick label的边框\底色\透明度\zorder
plt.show()
```

```python
#主次坐标轴任务
x = np.arange(0, 10, 0.1)
y1 = 0.05 * x**2
y2 = -1 *y1

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()    # mirror the ax1
#镜像y轴
ax1.plot(x, y1, color='green')
ax2.plot(x, y2, color='red')
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='gold')
ax2.set_ylabel('Y2 data', color='blue')

plt.show()
```
